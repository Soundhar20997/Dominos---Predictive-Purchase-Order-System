```python
import pandas as pd

# Load the Excel file
file_path = "Pizza_Sale.xlsx"  
df = pd.read_excel(file_path, sheet_name="pizza_sales")

# Convert date and time columns to datetime format
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["order_time"] = pd.to_datetime(df["order_time"], format="%H:%M:%S", errors="coerce").dt.time

# Handle missing values
df["pizza_name_id"].fillna("unknown", inplace=True)
df["total_price"].fillna(df["quantity"] * df["unit_price"], inplace=True)
df["pizza_category"].fillna("Unknown", inplace=True)
df["pizza_ingredients"].fillna("Not Available", inplace=True)
df["pizza_name"].fillna("Unknown Pizza", inplace=True)

# Drop duplicates
df.drop_duplicates(inplace=True)

# Save cleaned data
df.to_csv("Cleaned_Pizza_Sales.csv", index=False)

print("Data cleaning completed. Cleaned file saved as 'Cleaned_Pizza_Sales.csv'.")
```

Load and Merge Data

```python
import pandas as pd

# Load datasets
sales_data = pd.read_csv(r"C:\Soundhar\VS Code\project\Dominos\Cleaned_Pizza_Sales.csv")
ingredients_data = pd.read_csv(r"C:\Soundhar\VS Code\project\Dominos\Cleaned_Pizza_Ingredients.csv")

# Merge datasets on pizza_name_id
merged_data = sales_data.merge(ingredients_data, on="pizza_name_id", how="left")

# Save merged data
merged_data.to_csv("Merged_Pizza_Data.csv", index=False)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check missing values
print(merged_data.isnull().sum())

```

```python
merged_data["order_date"] = pd.to_datetime(merged_data["order_date"], format="%d-%m-%Y", errors="coerce")
```

```python
print(merged_data["order_date"].isna().sum())  # Check how many dates failed conversion
print(merged_data["order_date"].head())  # Display first few converted dates
```

```python
merged_data.dropna(subset=["order_date"], inplace=True)
```

```python
merged_data["order_date"].fillna(merged_data["order_date"].min(), inplace=True)
```

```python

# Aggregate sales by date
sales_trends = merged_data.groupby("order_date")["quantity"].sum()
```

Sales Trend Over Time

```python
plt.figure(figsize=(12, 6))
merged_data.groupby("order_date")["quantity"].sum().plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

Weekly & Monthly Sales Trends

```python
merged_data["month"] = merged_data["order_date"].dt.month
merged_data["day_of_week"] = merged_data["order_date"].dt.day_name()

# Monthly Sales
plt.figure(figsize=(10, 5))
sns.barplot(x=merged_data["month"], y=merged_data["quantity"], estimator=sum, palette="viridis")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Day of the Week Sales
plt.figure(figsize=(10, 5))
sns.barplot(x=merged_data["day_of_week"], y=merged_data["quantity"], estimator=sum, order=[
    "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.title("Sales by Day of the Week")
plt.xlabel("Day of Week")
plt.ylabel("Total Sales")
plt.show()
```

Distribution of Sales

```python
plt.figure(figsize=(8, 4))
sns.histplot(merged_data["quantity"], bins=30, kde=True)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.show()
```

Top Selling Pizzas

```python
top_pizzas = merged_data.groupby("pizza_name_x")["quantity"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_pizzas.index, y=top_pizzas.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Best Selling Pizzas")
plt.xlabel("Pizza Name")
plt.ylabel("Total Sales")
plt.show()
```
